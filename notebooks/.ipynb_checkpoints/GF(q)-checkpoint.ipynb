```julia
isacceptable(H, nedges) = isfullcore(permutedims(H)[1:end-1,:]) && (!has_multi_edges(H, nedges))
```

```julia
function maxsum(Q::Int, n::Int, R::Real; maxtrials=10^4, maxiter=10^4, rein=1e-3, tmax=3, gftab=gftables(Val(Q)),
        isacceptable=isacceptable, rng=MersenneTwister(abs(rand(Int))))
    A = cycle_code(Q, n, R, maxtrials=maxtrials, isacceptable=isacceptable, rng=rng)[1:end-1,:]
    s = rand(1:Q, n)
    minp = 2n; mindist = Inf
    for t in 1:tmax
        ms = ms_full_gfq(A, Q, s)
        for i in eachindex(ms.efield)
            ms.efield[i] = ms.efield[i] .+ 1e-5*randn(Q)
        end
        ε, iters = iteration_ms_quick!(ms, maxiter=maxiter, rein=rein, tol=1e-8, damp=0.0)
        x = argmax.(ms.belief)
        p = parity(ms, x)
        if p == 0
            mindist = distortion(x, s, Q)
            minp = 0
            break
        elseif p < minp
            B, indep = findbasis(A, Q, gftab=gftab)
            y = x .- 1
            fix_indep!(y, B, indep, Q)
            mindist = distortion(y .+ 1, s, Q)
            minp = p
        end
    end
    mindist
end
```

```julia
Qs = [2, 4, 8, 64]
Rs = 0.15:0.1:0.85
navg = 10
tmax = 5
d = [[fill(NaN,navg) for r in Rs] for q in Qs]
maxiter = 10^4
rein = 1e-3
n = 2*3*4*5*7*8
```

```julia
@telegram "GF(q)" for (i,Q) in pairs(Qs)
    println("### Q $i of ", length(Qs), " ###")
    gftabs = gftables(Val(Q))
    for (j,R) in pairs(Rs)
        println("---- Rate $j of ", length(Rs), " ###")
        Threads.@threads for k in 1:navg
            d[i][j][k] = maxsum(Q, Int(n÷log2(Q)), R; maxtrials=10^4, maxiter=maxiter, rein=rein,
                tmax=tmax, gftab=gftabs, rng=MersenneTwister(k))
        end
    end
    d_avg = mean.(d[i])
    d_std = std.(d[i]) ./ sqrt(navg)
    open("gf$(Q)_new.txt", "w") do io
       writedlm(io, [Rs d_avg d_std])
   end
end
```

```julia
function iteration_saveall!(bp::BPFull{F,SVector{Q,T}}; maxiter=10^3, tol=1e-12, 
        damp=0.0, rein=0.0, 
        update_f! = update_factor_bp!, update_v! = update_var_bp!,
        factor_neigs = [nonzeros(bp.X)[nzrange(bp.X, a)] for a = 1:size(bp.H,1)],
        gftab = gftables(Val(getQ(bp))), uaux=fill(zero(MVector{Q,T}),nfactors(bp)),
        parities=zeros(Int, maxiter), dist=zeros(maxiter), s=argmax.(bp.efield),
        callback=(it, ε, bp)->false) where {F,Q,T}

    Ht = permutedims(bp.H)

    ε = 0.0
    m,n = size(bp.H)
    for it = 1:maxiter
        ε = 0.0
        # for i = 1:size(bp.H,2)
        for i = rand(1:n, n÷3*2) 
            errv = update_v!(bp, i, damp=damp, rein=rein)
            errv == Inf && @warn "Contradiction found updating var $i"
            ε = max(ε, errv)
        end
        # for a = 1:size(bp.H,1)
        for a = rand(1:m, m÷3*2)
            errf = update_f!(bp, a, gftab..., factor_neigs[a], uaux=uaux[a], 
                damp=damp)
            ε = max(ε, errf)
        end
        x = argmax.(bp.belief)
        parities[it] = parity(bp, x, Ht)
        dist[it] = distortion(x,s,getQ(bp))
        callback(it, ε, bp) && return ε,it
        ε < tol && return ε, it
    end
    ε, maxiter
end

# alias for calling `iteration_saveall!` with maxsum updates
function iteration_ms_saveall!(ms::BPFull{F,SVector{Q,T}}; update_f! = update_factor_ms_quick!,
        update_v! = update_var_ms_quick!, kw...) where {F,Q,T}
    iteration_saveall!(ms; update_f! = update_f!, update_v! = update_v!, kw...)
end
```

## Varying the rate

```julia
include("../code/bp_full_gfq.jl")
include("../../telegram/notifications.jl")
```

```julia
Rs = 0.15:0.1:0.85
navg = 3
tmax = 3
d = [fill(NaN,navg) for r in Rs]
maxiter = 10^4
rein = 1e-3
damp = 0.0
k = 4
Q = 2^k
n = 3600
rng = MersenneTwister(0)
gftab = gftables(Val(Q));
```

```julia
@telegram "GF(q)" for (i,R) in pairs(Rs)
    println("### Rate $i of ", length(Rs), " ###")
    for j in 1:navg
        A = cycle_code(Q, n, R, maxtrials=10^4)[1:end-1,:]
        s = rand(1:Q, n)
        minp = 2n; mindist = Inf
        for t in 1:tmax
            ms = ms_full_gfq(A, Q, s)
            for i in eachindex(ms.efield)
                ms.efield[i] = ms.efield[i] .+ 1e-5*randn(Q)
            end
            ε, iters = iteration_ms_quick!(ms, maxiter=maxiter, rein=rein, tol=0, damp=damp)
            x = argmax.(ms.belief)
            p = parity(ms, x)
            if p == 0
                mindist = distortion(x, s, Q)
                minp = 0
                break
            elseif p < minp
                B, indep = findbasis(A, Q, gftab=gftab)
                y = x .- 1
                fix_indep!(y, B, indep, Q)
                mindist = distortion(y .+ 1, s, Q)
                minp = p
            end
        end
        println("Avg $j of $navg: min parity ", minp)
        d[i][j] = mindist
    end
end
```

```julia
using DelimitedFiles
M = readdlm("gf16.txt")
R_rs = M[:,1]
D_rs = M[:,2];
```

```julia
pl = plot_rdb()
scatter!(pl, Rs, mean.(d), label="GF($Q) - n = $n", ms=3, yerr=std.(d)./sqrt(navg), size=(400,400))
plot!(pl, R_rs, D_rs, label="RS GF(16)", ls=:dash)
plot(pl, size=(900,600), grid=false)
```

```julia
include("../code/rs.jl")
include("../code/bp_full.jl")
using Plots
```

```julia
RR = 0.001:0.01:0.999
triv = 0.5*(1 .- RR)
DD = 0.001:0.01:0.5
shann = 1 .+ DD.*log2.(DD) .+ (1 .- DD).*log2.(1 .- DD);
```

```julia
using Unzip

#k=mix of degrees l,l+1, d=2
Rrs1,Drs1=map(0.01:0.01:0.99) do r
    α=1-r
    k=floor(Int, 2/α)
    s=k+1-2/α
    K=[fill(0,k-1); s; 1-s]
    K .*= K .> 1e-10
    K ./ sum(K)
    r1,d,p = RS(K, [0,1]; N=100, tol=1e-10, maxiter=10000)
    @assert r ≈ r1
    r,d
end |>unzip 
```

```julia
#k=3, d = mix of degrees 2,3
Rrs2,Drs2=map(0.0:0.01:1/3) do r
    # 3-t=2t+3(1-t)=<d> = 3(1-r) -> t=3r
    t=3r
    r1,d,p = RS([0,0,1], [0,t,1-t]; N=20, tol=1e-10, maxiter=10000)
    @assert r ≈ r1
    r,d
end |>unzip
```

```julia
#k=3 , d=mix of degrees 1,2
Rrs3,Drs3=map(1/3:0.01:2/3) do r
    #2-t=t+2(1-t)=3(1-r) 
    t=3r-1
    r1,d,p = RS([0,0,1], [t,1-t]; N=20, tol=1e-10, maxiter=10000)
    @assert r ≈ r1
    r,d
end |> unzip
```

## Variables degree 2

```julia
Lambda = [0,1]
n1 = factorial(6)
nedges = 2n1
noise = 1e-10
R1 = 0.1:0.1:0.9
navg1 = 10
D1 = zeros(length(R1))
@showprogress for (i,r) in pairs(R1)
    α=1-r
    m = Int(round(α*n1,digits=10))
    k=floor(Int, 2/α)
    s=k+1-2/α
    K=[fill(0,k-1); s; 1-s]
    K .*= K .> 1e-10
    K ./ sum(K)
    cnt = 0
    for j in 1:navg1
        H = sparse(ldpc_matrix(n1, m, 2n1, Lambda, K)')
        s = rand((-1,1),n1)
        efield = [(ss+noise*randn(), -ss+noise*randn()) for ss in s]
        ms = BPFull(H, efield)
        nunsat = 1; dist = 0.0
        for q in 1:10
            ε, iters = iteration!(ms; maxiter=10^3, damp=0.5, rein=1e-3, 
                update_f! = update_factor_ms!, update_v! = update_var_ms!)
            nunsat, ovl, dist = performance(ms,s)
            if nunsat == 0 
                D1[i] += dist
                cnt += 1
                @show i, j, q
                break
            end
            fill!(ms.h, (0.0,0.0)); fill!(ms.u, (0.0,0.0)); fill!(ms.belief, (0.0,0.0)); ms.efield .= efield
        end
    end
    D1[i] /= cnt
end
```

## Variables degree 2 and 3

```julia
n2 = factorial(6)*2
K = [0,0,1]
f3s = 0.0:0.1:1
R2 = (1 .- f3s)/3
D2 = zeros(length(R2))
navg2 = 100

@showprogress for (i,f3) in pairs(f3s)
    Lambda = [0,1-f3,f3]
    R = (1 - f3)/3
    m = Int(round(n2*(1-R),digits=10))
    nedges = 3m
    cnt = 0
    for j in 1:navg2
        H = sparse(ldpc_matrix(n2, m, nedges, Lambda, K)')
        s = rand((-1,1),n2)
        efield = [(ss+noise*randn(), -ss+noise*randn()) for ss in s]
        ms = BPFull(H, efield)
        nunsat = 1; dist = 0.0
        for q in 1:1
            ε, iters = iteration!(ms; maxiter=10^3, damp=0.5, rein=1e-3, 
                update_f! = update_factor_ms!, update_v! = update_var_ms!)
            nunsat, ovl, dist = performance(ms,s)
            if nunsat == 0 
                D2[i] += dist
                cnt += 1
                @show i, j, q
                break
            end
            fill!(ms.h, (0.0,0.0)); fill!(ms.u, (0.0,0.0)); fill!(ms.belief, (0.0,0.0)); ms.efield .= efield
        end
    end 
    cnt==0 || (D2[i] /= cnt)
end
```

## Variables degree 1 and 2

```julia
n3 = factorial(6)
K = [0,0,1]
nedges = 2n3
noise = 1e-10
rr = 1/3:0.05:2/3
D3 = zeros(length(rr))
R3 = zeros(length(rr))
navg3 = 30
@showprogress for (i,r) in pairs(rr)
    t=3r-1
    Lambda = [t,1-t]
    α = sum(d*Lambda[d] for d in eachindex(Lambda)) / sum(k*K[k] for k in eachindex(K))
    m = Int(round(n3*α,digits=10))
    nedges = 3m
    cnt = 0
    for j in 1:navg3
        H = sparse(ldpc_matrix(n3, m, nedges, Lambda, K)')
        s = rand((-1,1),n3)
        efield = [(ss+noise*randn(), -ss+noise*randn()) for ss in s]
        ms = BPFull(H, efield)
        nunsat = 1; dist = 0.0
        for q in 1:1
            ε, iters = iteration!(ms; maxiter=10^3, damp=0.5, rein=1e-3, 
                update_f! = update_factor_ms!, update_v! = update_var_ms!)
            nunsat, ovl, dist = performance(ms,s)
            if nunsat == 0 
                D3[i] += dist
                cnt += 1
                @show i, j, q
                break
            end
            fill!(ms.h, (0.0,0.0)); fill!(ms.u, (0.0,0.0)); fill!(ms.belief, (0.0,0.0)); ms.efield .= efield
        end
    end
    D3[i] /= cnt
    R3[i] = 1 - α
end
```

## Plot

```julia
pl = plot(RR, triv, label="Trivial", xlabel="R", ylabel="D")
plot!(pl, shann, DD, label="Information bound")
plot!(pl, Rrs1, Drs1, ls=:dash, c=:gray, label="RS - Λ∼δ₂, K∼rδₗ+(1-r)δₗ₊₁")
plot!(pl, Rrs2, Drs2, label="RS - Λ∼rδ₂+(1-r)δ₃, K∼δ₃")
plot!(pl, Rrs3, Drs3, label="RS - Λ∼rδ₁+(1-r)δ₂, K∼δ₃")
plot!(pl, size=(500,400), dpi=100, lw=2);
```

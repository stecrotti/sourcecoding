# Distortion of unconverged Max-Sum instances
--------------
Three methods:
1. Assign by default the "naive compression" value $\frac{1-R}{2}$
2. Fix the independent variables to their value in the source and compute the others
3. Fix the independent variables to their value guessed by Max-Sum at the last iteration and compute the others

```julia
using JLD2
include("../code/headers.jl")
```

```julia
@load "../code/test/unconverged_distortion1.jld" sims
lt = @layout [a b;c d;e f]
p1 = plot(sims[1], allpoints=true, title="Naive compression")
p2 = plot(sims[1], title="Naive compression")
p3 = plot(sims[2], allpoints=true, title="Fix dependent vars to source")
p4 = plot(sims[2], title="Fix dependent vars to source")
p5 = plot(sims[3], allpoints=true, title="Fix dependent vars to MS output")
p6 = plot(sims[3], title="Fix dependent vars to MS output")

println("Distortions computed with the three methods.\nLeft: all points. Right: average")
plot(p1,p2,p3,p4,p5,p6, layout=lt, size=(800,800))
```

The last method works wonders!
